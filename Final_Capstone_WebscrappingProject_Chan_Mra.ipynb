```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime

web_page = "https://anycallmobilemm.com/product-category/apple/ipad/"
response = requests.get (web_page)
if response.status_code == 200:
  print ("Connection is successful!")

web_content = response.text
#print (web_content)

bsobj = BeautifulSoup (web_content, 'html.parser')
#print (bsobj)

product_tag_list = bsobj.find_all ('div', class_= 'product-element-bottom')
product_url_tag_list = bsobj.find_all ('a', class_= 'product-image-link')

name_list = []
price_list = []
product_url_list = []

for product_tag in product_tag_list:
  try:
    name_tag = product_tag.find ('h3', class_='wd-entities-title')
    name = name_tag.get_text ()
    price_tag = product_tag.find ('span', class_ = 'woocommerce-Price-amount amount')
    price = price_tag.get_text ().replace ("Ks",""). strip ()
    name_list.append (name)
    price_list.append (price)
  except Exception as e:
    print(f"Error processing product name or price: {e}")
    continue

for product_url_tag in product_url_tag_list:
  try:
    product_url = product_url_tag.get ('href')
    product_url_list.append (product_url)
  except Exception as e:
    print(f"Error processing product URL: {e}")
    continue

product_info_dic = {"Product Name": name_list,
                    "Product Price (MMK)": price_list,
                    "Product url": product_url_list}


df = pd.DataFrame (product_info_dic)
df.to_excel(f"Output_{datetime.now (). strftime ('%Y%m%d')}.xlsx", index = False)
```
